```python
import pandas as pd

import numpy as np
from scipy.optimize import curve_fit
```

```python
Year,Mean=np.loadtxt('annual_Mean_Temperatures.txt',unpack=True,usecols=(1,2),skiprows=1)
```

```python
#Checking the lowest mean annual temperature

Mean.min()
```

```python
#checking the maximum mean annual temperature
Mean.max()
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(Year,Mean,'.')
plt.xlabel('Year')
plt.ylabel('Annual Mean Temperatures')
```

```python
e=np.repeat(0.4,274)
e
```

```python
def line(x,a,b):
    return a*x+b
```

```python
popt,pcov=curve_fit(line,Mean,Year,sigma=e)
```

```python
popt
```

```python
pcov
```

```python
plt.errorbar(Mean,Year,yerr=e,fmt='.')
```

```python
plt.errorbar(Mean,Year,yerr=e,fmt='.')
xfine=np.linspace(0.,1.,10)
plt.plot(xfine,line(xfine,*popt),'r-')
```
